```python
def findDuplicate1(df, details_to_extract):
    temp = df[df['BMI'] != 200]
    temp_ids = temp.ID.index.tolist()
    og_ids = df.ID.index.tolist()
    #counts = dict()
    res = pd.DataFrame()
    res_id = [ temp_id for temp_id in temp_ids for og_id in og_ids if temp.loc[temp_id, 'ID'] == df.loc[og_id, 'ID'] and temp_id != og_id and  temp.loc[temp_id, 'Timestamp'] - df.loc[og_id, 'Timestamp'] <= timedelta(hours = 48)]
    print(res_id)
    res = pd.DataFrame(df, index = map(lambda x: x, res_id) )
    print(res)
    return res[details_to_extract]
```
