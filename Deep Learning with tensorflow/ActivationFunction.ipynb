```python
import math
def sigmoid(x):
    return 1/(1 + math.exp(-x))
```

```python
sigmoid(200), round(sigmoid(-200))
```

```python
def step(x):
    return 0 if x<0 else 1
```

```python
step(2), step(-2)
```

```python
def ReLU(x):
    return max(0,x)
```

```python
def leaky_ReLU(x):
    return max(0.1*x, x)
```

```python
import matplotlib.pyplot as plt
x=list()
y=list()
for i in range(-200,200,1):
    x.append(i)
    y.append(ReLU(i))
```

```python
plt.plot(x, y)
```

```python
x=list()
y=list()
for i in range(-200,200,1):
    x.append(i)
    y.append(leaky_ReLU(i))
```

```python
plt.plot(x,y)
```

```python
x=list()
y=list()
for i in range(-200,200,1):
    x.append(i)
    y.append(sigmoid(i))
plt.plot(x, y)
```

```python
x=list()
y=list()
for i in range(-200,200,1):
    x.append(i)
    y.append(step(i))
plt.plot(x, y)
```

```python
def tanh(x):
    return (math.exp(x)-math.exp(-x))/(math.exp(x)+math.exp(-x))
```

```python
x=list()
y=list()
for i in range(-200,200,1):
    x.append(i)
    y.append(tanh(i))
plt.plot(x, y)
```
